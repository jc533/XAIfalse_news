# drop duplicate

```python
# time series split
# drop time
# "text_y","text_x","replyId"
tmp_df.drop_duplicates(subset=["articleId"],keep=False,inplace=True)
tmp_df.drop_duplicates(subset=["text_x"],keep=False,inplace=True)
tmp_df.shape
```

```python
def re_url(url):
    t = str(url)
    regex = r'^https?:\/\/(www\.)?[-a-zA-Z0-9@:%._\+~#=]{1,256}\.[a-zA-Z0-9()]{1,6}\b([-a-zA-Z0-9()!@:%_\+.~#?&//=]*)$'
    if re.match(regex, t):
        return True
    else:
        return False
# re_url("www.google.com")
tmp_df["is_url"] = tmp_df["text_x"].apply(re_url)
tmp_df = tmp_df[tmp_df["is_url"]!=True]
print(tmp_df.shape)
# pd.DataFrame([i for i in range(len(i))],i)
# tmp_df
def have_url(url)->bool:
    t = str(url)
    regex = r'^https?:\/\/(www\.)?[-a-zA-Z0-9@:%._\+~#=]{1,256}\.[a-zA-Z0-9()]{1,6}\b([-a-zA-Z0-9()!@:%_\+.~#?&//=]*)$'
    if re.search(regex, t):
        return True
    else:
        return False
tmp_df["text_x"][tmp_df["text_x"].apply(have_url)]
```

```python
drop_col = [
    "is_url"
]
inplace = False
tmp_df = tmp_df.drop(drop_col,axis=1,inplace=inplace)
tmp_df.info()
tmp_df.head()
# article_and_replies.info()
```
