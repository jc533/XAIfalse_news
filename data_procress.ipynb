### structure
articleID,articleReplyID=> reply,reply_type


```python
articles = meta["articles"]
article_replies = meta["article_replies"]
replies = meta["replies"]
```

```python
article_with_replies = pd.merge(articles, article_replies, left_on='id', right_on='articleId', how='left')

# article_with_replies
```

```python
# classifier_traindata = article_with_replies[['text', 'replyType']]
# classifier_traindata
```

```python
article_and_replies = pd.merge(
    pd.merge( articles, article_replies, left_on='id', right_on='articleId', how='inner' ),
    replies, left_on='replyId', right_on='id', how='inner'
)
```

```python
# delte NOT_ARTICLE
print(article_and_replies.shape)
article_and_replies = article_and_replies[article_and_replies["type"]!="NOT_ARTICLE"]
print(article_and_replies.shape)
article_and_replies = article_and_replies[article_and_replies["articleType"]=="TEXT"]
print(article_and_replies.shape)
```

```python
drop_col = [
    "userIdsha256_x","userIdsha256_y",
    "lastRequestedAt",
    "appId_x","appId_y",
    "userIdsha256",
    "normalArticleReplyCount",
    "negativeFeedbackCount",
    "positiveFeedbackCount",
    "createdAt_x","createdAt_y",
    "updatedAt_x","updatedAt_y",
    "createdAt",
    "appId",
    "references"
    # "updatedAt",
]
inplace = False
tmp_df = article_and_replies.drop(drop_col,axis=1,inplace=inplace)
tmp_df.info()
tmp_df.head()
# article_and_replies.info()
```
